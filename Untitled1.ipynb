```python
import numpy as np

# Fungsi keanggotaan untuk Permintaan
def permintaan_turun(x):
    return max(min((3000 - x) / 3000, 1), 0)

def permintaan_tetap(x):
    return max(min((x - 2000) / 1000, (4000 - x) / 1000), 0)

def permintaan_naik(x):
    return max(min((x - 3000) / 3000, 1), 0)

# Fungsi keanggotaan untuk Persediaan
def persediaan_sedikit(x):
    return max(min((400 - x) / 400, 1), 0)

def persediaan_sedang(x):
    return max(min((x - 200) / 200, (600 - x) / 200), 0)

def persediaan_banyak(x):
    return max(min((x - 400) / 400, 1), 0)

# Fungsi keanggotaan untuk Produksi
def produksi_berkurang(z):
    return max(min((5000 - z) / 5000, 1), 0)

def produksi_bertambah(z):
    return max(min((z - 2000) / 3000, 1), 0)

# Inferensi berdasarkan aturan
def inferensi(permintaan, persediaan):
    rules = []
    
    # Rule 1
    μ1 = min(permintaan_turun(permintaan), persediaan_sedikit(persediaan))
    z1 = 7000 - μ1 * 5000  # Rule: Produksi Bertambah
    rules.append((μ1, z1))
    
    # Rule 2
    μ2 = min(permintaan_turun(permintaan), persediaan_sedang(persediaan))
    z2 = 5000 - μ2 * 5000  # Rule: Produksi Berkurang
    rules.append((μ2, z2))
    
    # Rule 3
    μ3 = min(permintaan_turun(permintaan), persediaan_banyak(persediaan))
    z3 = 5000 - μ3 * 5000  # Rule: Produksi Berkurang
    rules.append((μ3, z3))
    
    # Rule 4
    μ4 = min(permintaan_tetap(permintaan), persediaan_sedikit(persediaan))
    z4 = 7000 - μ4 * 5000  # Rule: Produksi Bertambah
    rules.append((μ4, z4))
    
    # Rule 5
    μ5 = min(permintaan_tetap(permintaan), persediaan_sedang(persediaan))
    z5 = 5000 - μ5 * 5000  # Rule: Produksi Berkurang
    rules.append((μ5, z5))
    
    # Rule 6
    μ6 = min(permintaan_tetap(permintaan), persediaan_banyak(persediaan))
    z6 = 5000 - μ6 * 5000  # Rule: Produksi Berkurang
    rules.append((μ6, z6))
    
    # Rule 7
    μ7 = min(permintaan_naik(permintaan), persediaan_sedikit(persediaan))
    z7 = 7000 - μ7 * 5000  # Rule: Produksi Bertambah
    rules.append((μ7, z7))
    
    # Rule 8
    μ8 = min(permintaan_naik(permintaan), persediaan_sedang(persediaan))
    z8 = 7000 - μ8 * 5000  # Rule: Produksi Bertambah
    rules.append((μ8, z8))
    
    # Rule 9
    μ9 = min(permintaan_naik(permintaan), persediaan_banyak(persediaan))
    z9 = 5000 - μ9 * 5000  # Rule: Produksi Berkurang
    rules.append((μ9, z9))
    
    return rules

# Defuzzification dengan metode Tsukamoto
def defuzzifikasi(rules):
    numerator = sum(μ * z for μ, z in rules)
    denominator = sum(μ for μ, _ in rules)
    return numerator / denominator if denominator != 0 else 0

# Contoh Input
permintaan = 3500  # Nilai permintaan
persediaan = 300   # Nilai persediaan

# Proses Inferensi
rules = inferensi(permintaan, persediaan)

# Hasil Defuzzifikasi
output_produksi = defuzzifikasi(rules)
print(f"Produksi (crisp output): {output_produksi}")
```
